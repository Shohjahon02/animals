```python
import plotly.graph_objects as go
from plotly.data import tips

df = tips()

summed_values = df.groupby(by="day", as_index=False).sum(numeric_only=True)
day_order_mapping = {"Thur": 0, "Fri": 1, "Sat": 2, "Sun": 3}
summed_values["order"] = summed_values["day"].apply(lambda day: day_order_mapping[day])
summed_values = summed_values.sort_values(by="order")

days_of_week = summed_values["day"].values
total_bills = summed_values["total_bill"].values
number_of_diners = summed_values["size"].values


fig = go.Figure(
    data=go.Bar(
        x=days_of_week,
        y=number_of_diners,
        name="Total number of diners",
        marker=dict(color="#4D3885"),
    )
)

fig.add_trace(
    go.Scatter(
        x=days_of_week,
        y=total_bills,
        yaxis="y2",
        name="Total bill amount",
        marker=dict(color="#8E92FE", size=10),
        line=dict(color='#8E92FE',
            width=4,
            dash='dash')
    )
)

fig.update_layout(
    legend=dict(orientation="h"),
    yaxis=dict(
        title=dict(text="Total number of diners"),
        side="left",
        range=[0, 250],
    ),
    yaxis2=dict(
        title=dict(text="Total bill amount"),
        side="right",
        range=[0, 2000],
        overlaying="y",
        tickmode="sync",
    ),
)
fig.update_layout(
    width=700,
    height=450
)

fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Indicator(
    value = 200,
    delta = {'reference': 160},
    gauge = {
        'axis': {'visible': False}},
    domain = {'row': 0, 'column': 0}))


fig.update_layout(
    width=450,
    height=450,
    grid = {'rows': 1, 'columns': 1, 'pattern': "independent"},
    template = {'data' : {'indicator': [{
        'mode' : "number+delta+gauge",
        'delta' : {'reference': 90}}]
                         }})
```

```python
xs = np.arange(10)
fig = go.FigureWidget()
fig.add_scatter(x=xs, y=np.sin(xs)+xs, fill='tozeroy')
fig.add_scatter(x=xs[[0,-1]], y=[11,11], fill='toself')
fig.layout = dict(xaxis=dict(linecolor='black', mirror=True),
                  yaxis=dict(zeroline=False, linecolor='black', mirror=True, range=[0,10]),
                  height=400, width=600, showlegend=False
                 )
fig.show()
```
